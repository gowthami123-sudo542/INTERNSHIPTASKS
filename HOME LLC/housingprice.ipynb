```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()

for col, value in hl.items():
    sns.distplot(value, ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

##Conclusion:By seeing the graph we come to know that data is not normalized.

```python
##let's check the outliers  by using boxplot

fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10)) ##used to create all the plots in one  plane 
index = 0
ax = ax.flatten()

for col, value in hl.items():
    sns.boxplot(y=col, data=hl, ax=ax[index])   ##boxplot is used to check mainly for outliers
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

##Apply min-max normalization to normalize the features so it reduces the  error rate

```python
cols = ['crim', 'zn', 'tax', 'black']
for col in cols:
  
    minimum = min(hl[col])
    maximum = max(hl[col])
    hl[col] = (hl[col] - minimum) / (maximum - minimum)
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()

for col, value in hl.items():
    sns.distplot(value, ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

##conclusion:it does not affect much because we have different scales so let's apply standardization techinque

```python
# standardization
from sklearn import preprocessing
scalar = preprocessing.StandardScaler()
scaled_cols = scalar.fit_transform(hl[cols])
scaled_cols = pd.DataFrame(scaled_cols, columns=cols)
scaled_cols.head()
```

```python
##let's transform these values into original dataset
for col in cols:
    hl[col] = scaled_cols[col]
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()

for col, value in hl.items():
    sns.distplot(value, ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
corr = hl.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

##Conclusion:Here tax,rad are highly correlated to each other,so let's ignore one 

```python
x = hl.drop(columns=['medv', 'rad'], axis=1)
y = hl['medv']
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error
def train(model, x, y):
    ## split the dataset
    x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
    model.fit(x_train, y_train)
    
    ## predict the training set
    pred = model.predict(x_test)
    
    ## perform cross-validation
    cv_score = cross_val_score(model, x, y, scoring='neg_mean_squared_error', cv=5)
    cv_score = np.abs(np.mean(cv_score))
    
    print("Model Report")
    print("MSE:",mean_squared_error(y_test, pred))
    print('CV Score:', cv_score)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(normalize=True)
train(model, x, y)
coef = pd.Series(model.coef_, x.columns).sort_values()
coef.plot(kind='bar', title='Model Coefficients')
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
train(model, x, y)
coef = pd.Series(model.feature_importances_, x.columns).sort_values(ascending=False)
coef.plot(kind='bar', title='Feature Importance')
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
train(model, x, y)
coef = pd.Series(model.feature_importances_, x.columns).sort_values(ascending=False)
coef.plot(kind='bar', title='Feature Importance')
```

```python
##conclusion:By using RandomForestRegressor we got MSE:10.33 which is less compared to LinearRegression algorthim.
```
